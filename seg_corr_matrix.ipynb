```python
df.columns = ['user_id', 'city_level', 'year_segment', 'android_flag', 'conversion_timing',
       'unsign_times', 'retained_flag', 'upgrade_flag',
       'first_unsign_timing', 'last_unsign_timing', 'course_segment',
       'leads_age', 'new_third_level',
       'resource_level', 'group_tag', 'days_lapsed',
       'train_study_rate', 'train_finish_rate', 'train_same_finish',
       'previous_leads', 'year_course_cnt', 'year_start_rate',
       'year_finish_rate', 'year_period_finish', 'last_start_rate',
       'last_finish_rate', 'last_period_finish']
```

```python
import numpy as np
```

```python
df['unsign_flag'] = np.where(df['unsign_times']> 0, 1, 0)
df['prev_lead_flag'] = np.where(df['previous_leads']> 0, 1, 0)
```

```python
df.describe()
```

```python
secnario_1 = df[['city_level', 'conversion_timing', 'android_flag',
       'unsign_times', 'retained_flag', 
       'first_unsign_timing', 'last_unsign_timing',
       'leads_age',
       'resource_level', 'days_lapsed',
       'train_study_rate', 'train_finish_rate', 'train_same_finish',
       'previous_leads', 'year_course_cnt', 'year_start_rate',
       'year_finish_rate', 'year_period_finish', 'last_start_rate',
       'last_finish_rate', 'last_period_finish', 'unsign_flag', 'prev_lead_flag']]
```

```python
matrix = secnario_1.corr()
print(matrix)
```

```python
secnario_2 = df[['city_level', 'conversion_timing',
       'unsign_times', 'retained_flag', 
       'last_unsign_timing',
       'leads_age',
       'resource_level', 'days_lapsed',
       'train_finish_rate', 
       'previous_leads', 'year_course_cnt',
       'last_finish_rate']]
```

```python
matrix = secnario_2.corr().round(3)
print(matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(matrix, annot=True, cmap= 'vlag', linewidths=.2)
plt.show()
```
